Modelo de Machine Learning para prever o indíce de satisfação dos clientes.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
dados = pd.read_csv('dadoshistoricos.csv')
type(dados)
```

```python
dados.head(5)
```

Removendo a coluna ID

```python
dados = dados.drop('ID', axis=1)
dados.shape
```

```python
dados.head(10)
```

Separando a variável TARGET que seria o indíce de satisfação do cliente

0 -> não

1 -> sim

```python
TARGET = dados[['TARGET']]
```

Criando uma lista com as colunas do objeto.

```python
colunas = dados.columns[0:369].tolist()
colunas
```

Criando um array dos dados.

```python
array = dados.values
type(array)
```

Separando o array em componentes de input (x) da vairável TARGET. Fazendo a normalização para deixar todos os dados na mesma escala.

```python
x = array[:, 0:369]

scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(x)
```

Transformando o array em DataFrame.

```python
dadosNormalizados = pd.DataFrame(rescaledX, columns = colunas)
dadosNormalizados.shape
```

Reduzindo a dimensionalidade.

```python
pca = PCA(n_components = 30)
dados = pca.fit_transform(dadosNormalizados)
```

Convertendo o objeto PCA para um DataFrame.

```python
dados = pd.DataFrame(dados)
dados.head(5)
```

Concatenando a variável TARGET ao DataFrame.

```python
dados = pd.concat([dados, TARGET], axis=1)
dados
```

Dados de treino e teste para passar no nosso modelo preditivo. Nos dados de treino iremos fazer uma amostra com 60% dos dados e no teste iremos pegar os 40% restantes de dados.

```python
dados_treino = dados.sample(frac=0.60, random_state=0)
dados_teste = dados.drop(dados_treino.index)
```

Gerando as variáveis preditoras e variável target.

```python
x_treino = dados_treino.iloc[:, 0:30]
y_treino = dados_treino.iloc[:, 30:31]

x_teste = dados_teste.iloc[:, 0:30]
y_teste = dados_teste.iloc[:, 30:31]
```

```python
dados_treino.count()
```

```python
dados_teste.count()
```

Classificador Random Forest

```python
classificador = RandomForestClassifier(max_depth=5, random_state=0)
classificador.fit(x_treino, y_treino)

# executando as previsoes
previsoes = classificador.predict(x_teste)
```

Verificando a Confusion Matrix e acurácia.

```python
print('Acuracia: %s' % accuracy_score(y_teste, previsoes))
```
